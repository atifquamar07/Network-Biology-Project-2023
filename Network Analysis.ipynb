```python
import networkx as nx
import pandas as pd

file_path = "datasets/test_gene_data.txt"

data = pd.read_csv(file_path, sep='\t')

genes = data['Symbol'].tolist()
expression_data = data.drop('Symbol', axis=1).values

correlation_matrix = pd.DataFrame(expression_data).T.corr()

G = nx.Graph()
G.add_nodes_from(genes)
for i in range(len(genes)):
    for j in range(i + 1, len(genes)):
        G.add_edge(genes[i], genes[j], weight=correlation_matrix.iloc[i, j])
```

```python
import matplotlib.pyplot as plt

pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, font_size=8, node_size=100, font_color='black', alpha=0.6)
plt.show()
```

```python
degree_centrality = nx.degree_centrality(G)

betweenness_centrality = nx.betweenness_centrality(G)
```
